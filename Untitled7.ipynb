```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Load the dataset
movies = pd.read_csv('/content/movies.csv')  # Adjust the path as necessary

# Check the first few rows of the dataset
print(movies.head())

# Preprocess the data
# Combine the title and genres into a single string for each movie
movies['tags'] = movies['name'] + ' ' + movies['genre']

# Create a CountVectorizer to convert text data into a matrix of token counts
cv = CountVectorizer(max_features=10000, stop_words='english')
vector = cv.fit_transform(movies['tags']).toarray()

# Calculate the cosine similarity matrix
similarity = cosine_similarity(vector)

# Function to recommend movies
def recommend(movie_title):
    try:
        # Get the index of the movie that matches the title
        index = movies[movies['name'] == movie_name].index[0]

        # Get the pairwise similarity scores of all movies with that movie
        distances = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda x: x[1])

        # Get the titles of the top 5 similar movies
        recommended_movies = []
        for i in distances[1:6]:  # Skip the first one as it is the same movie
            recommended_movies.append(movies.iloc[i[0]]['title'])

        return recommended_movies
    except IndexError:
        return "Movie not found!"

# Example usage
if __name__ == "__main__":
    movie_name = input("Enter a movie title: ")
    recommendations = recommend(movie_name)
    print("Recommended movies:")
    for movie in recommendations:
        print(movie)
```
